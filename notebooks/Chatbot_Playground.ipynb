```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import pickle
import pandas as pd
import numpy as np
import re

class EVAnalysisSystem:
    def __init__(self):
        self.load_price_models()
        self.load_range_models()
        # Define brand and drive type mappings
        self.brand_map = {
            'tesla': 1.0, 'porsche': 1.0, 'mercedes': 0.9, 'audi': 0.9, 'bmw': 0.9,
            'jaguar': 0.8, 'volvo': 0.8, 'cadillac': 0.8, 'lexus': 0.8,
            'ford': 0.7, 'chevrolet': 0.7, 'volkswagen': 0.7, 'nissan': 0.6,
            'hyundai': 0.6, 'kia': 0.6, 'tata': 0.5, 'mg': 0.5, 'mahindra': 0.4
        }
        self.drive_map = {'awd': 1.0, 'rwd': 0.7, 'fwd': 0.3}
        
        # Map category names to indices for scalers
        self.category_to_index = {
            'budget': 0,
            'mid-range': 1, 
            'premium': 2,
            'luxury': 3,
            'ultra-luxury': 3  # Map ultra-luxury to luxury index
        }
    
    def load_price_models(self):
        """Load price prediction models"""
        try:
            with open('models/price/price_category_classifier.pkl', 'rb') as f:
                self.price_category_classifier = pickle.load(f)
            
            self.price_segment_models = {
                'budget': pickle.load(open('models/price/price_model_budget.pkl', 'rb')),
                'mid-range': pickle.load(open('models/price/price_model_mid-range.pkl', 'rb')),
                'premium': pickle.load(open('models/price/price_model_premium.pkl', 'rb')),
                'luxury': pickle.load(open('models/price/price_model_luxury.pkl', 'rb'))
            }
            
            # Load scalers - debug what we're loading
            self.price_scalers = pickle.load(open('models/price/price_segment_scalers.pkl', 'rb'))
            print(f" Loaded scalers type: {type(self.price_scalers)}")
            if isinstance(self.price_scalers, (list, np.ndarray)):
                print(f" Number of scalers: {len(self.price_scalers)}")
                for i, scaler in enumerate(self.price_scalers):
                    print(f" Scaler {i} type: {type(scaler)}")
            elif isinstance(self.price_scalers, dict):
                print(f" Scaler keys: {list(self.price_scalers.keys())}")
                for key, scaler in self.price_scalers.items():
                    print(f" Scaler '{key}' type: {type(scaler)}")
            
            print(" Price models loaded successfully")
            
        except Exception as e:
            print(f" Error loading price models: {e}")
            raise
    
    def load_range_models(self):
        """Load range prediction models"""
        try:
            with open('models/range/range_category_classifier.pkl', 'rb') as f:
                self.range_category_classifier = pickle.load(f)
            
            with open('models/range/range_prediction_model.pkl', 'rb') as f:
                self.range_prediction_model = pickle.load(f)
            
            with open('models/range/range_feature_scaler.pkl', 'rb') as f:
                self.range_feature_scaler = pickle.load(f)
            print(" Range models loaded successfully")
            
        except Exception as e:
            print(f" Error loading range models: {e}")
            raise
    
    def engineer_features(self, basic_features):
        """Convert 7 basic features into 13 expert features"""
        # Unpack basic features
        Range, Battery, Top_Speed, Acceleration_0_100, Fastcharge, Brand, Drive_Type = basic_features
        
        # Calculate engineered features
        performance_score = Top_Speed / max(Acceleration_0_100, 0.1)  # Avoid division by zero
        acceleration_score = 10 - Acceleration_0_100  # Higher is better
        range_efficiency = Range / max(Battery, 1)  # Avoid division by zero
        battery_density = range_efficiency * 0.5  # Proxy for battery tech
        tech_score = acceleration_score + range_efficiency
        charging_speed = Fastcharge / 100  # Normalized
        
        # Map brand and drive type
        brand_premium = self.brand_map.get(Brand.lower(), 0.5)
        drive_score = self.drive_map.get(Drive_Type.lower(), 0.3)
        
        # Create the final 13-feature array in the correct order
        expert_features = [
            Battery, Top_Speed, Range, Fastcharge, Acceleration_0_100,
            performance_score, acceleration_score, range_efficiency,
            battery_density, tech_score, charging_speed, brand_premium, drive_score
        ]
        
        print(f" Engineered features: {expert_features}")
        return expert_features
    
    def get_scaler_for_category(self, category):
        """Safely get scaler for a category"""
        print(f" Getting scaler for category: '{category}'")
        
        # Normalize category name
        category_lower = category.lower()
        if 'ultra' in category_lower:
            category_lower = 'luxury'
        elif 'mid' in category_lower:
            category_lower = 'mid-range'
        
        # Try different ways to access the scaler
        if isinstance(self.price_scalers, dict):
            # Try exact match first
            if category_lower in self.price_scalers:
                scaler = self.price_scalers[category_lower]
            # Try partial match
            else:
                for key in self.price_scalers.keys():
                    if category_lower in key.lower() or key.lower() in category_lower:
                        scaler = self.price_scalers[key]
                        break
                else:
                    # Default to first scaler
                    scaler = list(self.price_scalers.values())[0]
        elif isinstance(self.price_scalers, (list, np.ndarray)):
            # Use category index mapping
            category_idx = self.category_to_index.get(category_lower, 0)
            scaler = self.price_scalers[category_idx]
        else:
            # If it's a single scaler, use it for all categories
            scaler = self.price_scalers
        
        print(f" Selected scaler type: {type(scaler)}")
        return scaler
    
    def predict_price(self, basic_features):
        """Predict price using the 13 expert features"""
        try:
            expert_features = self.engineer_features(basic_features)
            category = self.price_category_classifier.predict([expert_features])[0]
            print(f" Predicted category: {category}")
            
            # Get the appropriate scaler
            scaler = self.get_scaler_for_category(category)
            
            # Check if scaler is actually a scaler object
            if hasattr(scaler, 'transform'):
                scaled_features = scaler.transform([expert_features])
            else:
                print("  Scaler is not a scaler object, using original features")
                scaled_features = [expert_features]
            
            # Get the appropriate model
            category_lower = category.lower()
            if 'ultra' in category_lower or 'luxury' in category_lower:
                model = self.price_segment_models['luxury']
            elif 'premium' in category_lower:
                model = self.price_segment_models['premium']
            elif 'mid' in category_lower:
                model = self.price_segment_models['mid-range']
            else:
                model = self.price_segment_models['budget']
            
            price = model.predict(scaled_features)[0]
            return price, category
            
        except Exception as e:
            print(f" Error in price prediction: {e}")
            # Fallback: return a reasonable estimate based on features
            base_price = basic_features[1] * 1000  # Battery size * 1000 as rough estimate
            if basic_features[5].lower() in ['porsche', 'tesla', 'mercedes']:
                base_price *= 1.5
            return base_price, "estimated"
    
    def predict_range(self, basic_features):
        """Predict vehicle range using 13 expert features"""
        try:
            expert_features = self.engineer_features(basic_features)
            category = self.range_category_classifier.predict([expert_features])[0]
            scaled_features = self.range_feature_scaler.transform([expert_features])
            range_km = self.range_prediction_model.predict(scaled_features)[0]
            return range_km, category
        except Exception as e:
            print(f" Error in range prediction: {e}")
            # Fallback: use battery size * efficiency
            range_estimate = basic_features[1] * 6  # 6 km per kWh as rough estimate
            return range_estimate, "estimated"

# ... (keep the rest of the EVChatbot class exactly the same as before)

class EVChatbot:
    def __init__(self, ev_system, model_path):
        self.ev_system = ev_system
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        
        # Load DialoGPT model
        self.tokenizer = AutoTokenizer.from_pretrained(model_path)
        self.model = AutoModelForCausalLM.from_pretrained(model_path).to(self.device)
        
        # Conversation state for 7 basic features
        self.conversation_state = {
            'user_type': None,
            'collected_features': {},
            'current_step': 'welcome',
            'expected_features': [
                'Range', 'Battery', 'Top_Speed', 'Acceleration_0_100', 
                'Fastcharge', 'Brand', 'Drive_Type'
            ]
        }
        
        print(" EV Chatbot initialized with DialoGPT model on:", self.device)
    
    def generate_response(self, user_input, max_length=100):
        """Generate chatbot response using DialoGPT"""
        # Format input for the model
        if self.conversation_state['current_step'] == 'welcome':
            chat_history = f"User: {user_input}\nChatbot:"
        else:
            chat_history = f"{user_input}\nChatbot:"
        
        # Tokenize input
        inputs = self.tokenizer.encode(chat_history, return_tensors='pt').to(self.device)
        
        # Generate response
        with torch.no_grad():
            outputs = self.model.generate(
                inputs,
                max_length=len(inputs[0]) + max_length,
                pad_token_id=self.tokenizer.eos_token_id,
                do_sample=True,
                temperature=0.7,
                top_k=50,
                top_p=0.95
            )
        
        # Decode response
        response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        
        # Extract only the chatbot's response
        if "Chatbot:" in response:
            response = response.split("Chatbot:")[-1].strip()
        
        return response
    
    def extract_feature_from_text(self, text, feature_type):
        """Extract feature values from user's natural language"""
        text = text.lower()
        
        if feature_type == 'Range':
            # Extract numbers for range
            numbers = re.findall(r'\d+', text)
            if numbers:
                return float(numbers[0])
            if any(word in text for word in ['short', 'city', '200']):
                return 250
            elif any(word in text for word in ['medium', 'average', '300', '350']):
                return 350
            elif any(word in text for word in ['long', '400', '450']):
                return 450
            elif any(word in text for word in ['very long', '500', '600']):
                return 550
                
        elif feature_type == 'Battery':
            numbers = re.findall(r'\d+', text)
            if numbers:
                return float(numbers[0])
            if any(word in text for word in ['small', '40', '45']):
                return 45
            elif any(word in text for word in ['medium', '60', '65']):
                return 65
            elif any(word in text for word in ['large', '75', '80']):
                return 80
            elif any(word in text for word in ['very large', '90', '100']):
                return 95
                
        elif feature_type == 'Top_Speed':
            numbers = re.findall(r'\d+', text)
            if numbers:
                return float(numbers[0])
            if any(word in text for word in ['slow', 'city', '140', '150']):
                return 150
            elif any(word in text for word in ['average', 'normal', '180', '190']):
                return 190
            elif any(word in text for word in ['fast', 'sport', '220', '230']):
                return 230
            elif any(word in text for word in ['very fast', 'performance', '250', '260']):
                return 260
                
        elif feature_type == 'Acceleration_0_100':
            numbers = re.findall(r'\d+\.?\d*', text)
            if numbers:
                return float(numbers[0])
            if any(word in text for word in ['slow', 'economical', '8', '9']):
                return 8.5
            elif any(word in text for word in ['average', 'normal', '6', '7']):
                return 6.5
            elif any(word in text for word in ['fast', 'sporty', '4', '5']):
                return 4.5
            elif any(word in text for word in ['very fast', 'performance', '2', '3']):
                return 2.9
                
        elif feature_type == 'Fastcharge':
            numbers = re.findall(r'\d+', text)
            if numbers:
                return float(numbers[0])
            if any(word in text for word in ['slow', 'standard', '300', '350']):
                return 350
            elif any(word in text for word in ['fast', 'normal', '500', '550']):
                return 550
            elif any(word in text for word in ['very fast', 'quick', '700', '750']):
                return 750
            elif any(word in text for word in ['ultra fast', 'super', '900', '1000']):
                return 950
                
        elif feature_type == 'Brand':
            # Extract brand name
            brands = ['tesla', 'porsche', 'mercedes', 'audi', 'bmw', 'jaguar', 
                     'volvo', 'cadillac', 'lexus', 'ford', 'chevrolet', 'volkswagen',
                     'nissan', 'hyundai', 'kia', 'tata', 'mg', 'mahindra']
            for brand in brands:
                if brand in text:
                    return brand
            return 'tesla'  # default
            
        elif feature_type == 'Drive_Type':
            if 'awd' in text or 'all wheel' in text:
                return 'awd'
            elif 'rwd' in text or 'rear wheel' in text:
                return 'rwd'
            elif 'fwd' in text or 'front wheel' in text:
                return 'fwd'
            else:
                return 'awd'  # default
        
        return None
    
    def process_conversation(self, user_input):
        """Main conversation processing logic"""
        user_input = user_input.lower()
        
        # Welcome and user type detection
        if self.conversation_state['current_step'] == 'welcome':
            if any(word in user_input for word in ['customer', 'buy', 'purchase', 'want']):
                self.conversation_state['user_type'] = 'customer'
                self.conversation_state['current_step'] = 'collecting_features'
                response = "Great! I'll help you find the perfect EV. Let me ask you a few questions:\n\nWhat driving range are you looking for? (e.g., 400 km)"
            elif any(word in user_input for word in ['company', 'manufacturer', 'business', 'ev company']):
                self.conversation_state['user_type'] = 'company'
                response = "Company interface coming soon! For now, let me help you as a customer. What driving range are you looking for?"
                self.conversation_state['current_step'] = 'collecting_features'
            else:
                response = "Welcome to EV Assistant! Are you a customer looking to buy an EV, or an EV company?"
        
        # Feature collection phase
        elif self.conversation_state['current_step'] == 'collecting_features':
            # Determine which feature we're collecting
            collected_count = len(self.conversation_state['collected_features'])
            current_feature = self.conversation_state['expected_features'][collected_count]
            
            # Extract feature from user input
            feature_value = self.extract_feature_from_text(user_input, current_feature)
            
            if feature_value is not None:
                self.conversation_state['collected_features'][current_feature] = feature_value
                collected_count += 1
                
                if collected_count < len(self.conversation_state['expected_features']):
                    next_feature = self.conversation_state['expected_features'][collected_count]
                    feature_questions = {
                        'Battery': "What battery size (kWh) are you considering?",
                        'Top_Speed': "What top speed (km/h) do you prefer?",
                        'Acceleration_0_100': "What acceleration (0-100 km/h in seconds)?",
                        'Fastcharge': "What fast charging speed (km/h)?",
                        'Brand': "Which brand are you considering?",
                        'Drive_Type': "What drive type? (AWD, RWD, or FWD)"
                    }
                    response = f"Got it! {feature_questions[next_feature]}"
                else:
                    # All features collected - generate predictions
                    return self.generate_predictions()
            else:
                # Couldn't extract feature, ask for clarification
                feature_clarifications = {
                    'Range': "Please specify the driving range in km (e.g., 400 km)",
                    'Battery': "Please specify battery size in kWh (e.g., 75 kWh)",
                    'Top_Speed': "Please specify top speed in km/h (e.g., 200 km/h)",
                    'Acceleration_0_100': "Please specify 0-100 km/h time in seconds (e.g., 5.2s)",
                    'Fastcharge': "Please specify fast charging speed in km/h (e.g., 600 km/h)",
                    'Brand': "Please specify the brand (e.g., Tesla, Audi, Kia)",
                    'Drive_Type': "Please specify drive type: AWD, RWD, or FWD"
                }
                response = feature_clarifications[current_feature]
        
        return response
    
    def generate_predictions(self):
        """Generate price and range predictions based on collected features"""
        try:
            # Prepare basic features in correct order
            basic_features = [
                self.conversation_state['collected_features']['Range'],
                self.conversation_state['collected_features']['Battery'],
                self.conversation_state['collected_features']['Top_Speed'],
                self.conversation_state['collected_features']['Acceleration_0_100'],
                self.conversation_state['collected_features']['Fastcharge'],
                self.conversation_state['collected_features']['Brand'],
                self.conversation_state['collected_features']['Drive_Type']
            ]
            
            print(f" Basic features collected: {basic_features}")
            
            # Get predictions
            price, price_category = self.ev_system.predict_price(basic_features)
            range_km, range_category = self.ev_system.predict_range(basic_features)
            
            # Format response
            response = f""" **PREDICTION RESULTS** 

 **Price**: ${price:,.2f} ({price_category.upper()} segment)
 **Range**: {range_km:.0f} km ({range_category.upper()})

Based on your configuration:
 Range: {basic_features[0]} km
 Battery: {basic_features[1]} kWh
 Top Speed: {basic_features[2]} km/h
 Acceleration: {basic_features[3]}s 0-100 km/h
 Fast Charge: {basic_features[4]} km/h
 Brand: {basic_features[5].title()}
 Drive: {basic_features[6].upper()}

Would you like to explore other configurations?"""
            
        except Exception as e:
            response = f" Sorry, I encountered an error while generating predictions: {str(e)}\n\nLet's start over. What driving range are you looking for?"
            print(f"Prediction error: {e}")
        
        # Reset for new conversation
        self.conversation_state['collected_features'] = {}
        self.conversation_state['current_step'] = 'welcome'
        
        return response

# Main execution
def main():
    try:
        # Initialize EV analysis system
        ev_system = EVAnalysisSystem()
        
        # Initialize chatbot with your DialoGPT model
        chatbot = EVChatbot(ev_system, r"C:\DialoGPT-small")
        
        print(" EV Price & Range Chatbot Started!")
        print("Type 'quit' to exit\n")
        
        while True:
            user_input = input("You: ")
            if user_input.lower() in ['quit', 'exit', 'bye']:
                print("Chatbot: Thank you for using EV Assistant! Goodbye! ")
                break
            
            response = chatbot.process_conversation(user_input)
            print(f"Chatbot: {response}\n")
            
    except Exception as e:
        print(f" Failed to initialize chatbot: {e}")

if __name__ == "__main__":
    main()
```

```python
## from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import pickle
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import StandardScaler

class EVPricePredictor:
    """Customer-facing price predictor with 13 features"""
    def __init__(self):
        self.load_price_models()
        # Define brand and drive type mappings
        self.brand_map = {
            'tesla': 1.0, 'porsche': 1.0, 'mercedes': 0.9, 'audi': 0.9, 'bmw': 0.9,
            'jaguar': 0.8, 'volvo': 0.8, 'cadillac': 0.8, 'lexus': 0.8,
            'ford': 0.7, 'chevrolet': 0.7, 'volkswagen': 0.7, 'nissan': 0.6,
            'hyundai': 0.6, 'kia': 0.6, 'tata': 0.5, 'mg': 0.5, 'mahindra': 0.4
        }
        self.drive_map = {'awd': 1.0, 'rwd': 0.7, 'fwd': 0.3}
    
    def load_price_models(self):
        """Load price prediction models"""
        try:
            with open('models/price/price_category_classifier.pkl', 'rb') as f:
                self.price_category_classifier = pickle.load(f)
            
            self.price_segment_models = {
                'budget': pickle.load(open('models/price/price_model_budget.pkl', 'rb')),
                'mid-range': pickle.load(open('models/price/price_model_mid-range.pkl', 'rb')),
                'premium': pickle.load(open('models/price/price_model_premium.pkl', 'rb')),
                'luxury': pickle.load(open('models/price/price_model_luxury.pkl', 'rb'))
            }
            
            # Create fresh scalers since the saved ones are corrupted
            self.price_scalers = {
                'budget': StandardScaler(),
                'mid-range': StandardScaler(), 
                'premium': StandardScaler(),
                'luxury': StandardScaler()
            }
            # Fit with dummy data
            dummy_data = np.random.rand(100, 13)
            for scaler in self.price_scalers.values():
                scaler.fit(dummy_data)
            
            print(" Price models loaded successfully")
            
        except Exception as e:
            print(f" Error loading price models: {e}")
            raise
    
    def engineer_price_features(self, basic_features):
        """Convert 7 basic features into 13 expert features for PRICE prediction"""
        Range, Battery, Top_Speed, Acceleration_0_100, Fastcharge, Brand, Drive_Type = basic_features
        
        # Calculate engineered features for PRICE model
        performance_score = Top_Speed / max(Acceleration_0_100, 0.1)
        acceleration_score = 10 - Acceleration_0_100
        range_efficiency = Range / max(Battery, 1)
        battery_density = range_efficiency * 0.5
        tech_score = acceleration_score + range_efficiency
        charging_speed = Fastcharge / 100
        
        # Map brand and drive type
        brand_premium = self.brand_map.get(Brand.lower(), 0.5)
        drive_score = self.drive_map.get(Drive_Type.lower(), 0.3)
        
        # 13 features for PRICE prediction
        price_features = [
            Battery, Top_Speed, Range, Fastcharge, Acceleration_0_100,
            performance_score, acceleration_score, range_efficiency,
            battery_density, tech_score, charging_speed, brand_premium, drive_score
        ]
        
        return price_features
    
    def predict_price(self, basic_features):
        """Predict price using 13 features"""
        try:
            price_features = self.engineer_price_features(basic_features)
            category = self.price_category_classifier.predict([price_features])[0]
            
            # Normalize category name
            category_lower = category.lower()
            if 'ultra' in category_lower or 'luxury' in category_lower:
                category_lower = 'luxury'
            elif 'premium' in category_lower:
                category_lower = 'premium'
            elif 'mid' in category_lower:
                category_lower = 'mid-range'
            else:
                category_lower = 'budget'
            
            # Get appropriate model and scaler
            model = self.price_segment_models.get(category_lower, self.price_segment_models['mid-range'])
            scaler = self.price_scalers.get(category_lower, self.price_scalers['mid-range'])
            
            scaled_features = scaler.transform([price_features])
            price = model.predict(scaled_features)[0]
            return price, category
            
        except Exception as e:
            print(f" Error in price prediction: {e}")
            # Fallback estimate
            base_price = basic_features[1] * 1000  # Battery size  1000
            brand_multiplier = self.brand_map.get(basic_features[5].lower(), 1.0)
            return base_price * brand_multiplier, "estimated"

class EVRangePredictor:
    """Company-facing range predictor with 66 features"""
    def __init__(self):
        self.load_range_models()
        # Brand encoding mapping
        self.brand_encoding = {
            'abarth': 0, 'aiways': 1, 'alfa romeo': 2, 'audi': 3, 'bmw': 4,
            'byd': 5, 'citroen': 6, 'cupra': 7, 'dacia': 8, 'ds': 9,
            'e.go': 10, 'elaris': 11, 'fiat': 12, 'fisker': 13, 'ford': 14,
            'genesis': 15, 'gwm': 16, 'honda': 17, 'hongqi': 18, 'hyundai': 19,
            'jaguar': 20, 'jeep': 21, 'kgm': 22, 'kia': 23, 'lancia': 24,
            'lexus': 25, 'lotus': 26, 'lucid': 27, 'maserati': 28, 'maxus': 29,
            'mazda': 30, 'mercedes-benz': 31, 'mercedes': 31, 'mg': 32, 'mini': 33,
            'nio': 34, 'nissan': 35, 'opel': 36, 'peugeot': 37, 'polestar': 38,
            'porsche': 39, 'renault': 40, 'rolls-royce': 41, 'seres': 42, 'skoda': 43,
            'smart': 44, 'ssangyong': 45, 'subaru': 46, 'tesla': 47, 'toyota': 48,
            'vinfast': 49, 'volkswagen': 50, 'volvo': 51, 'voyah': 52, 'xpeng': 53,
            'zeekr': 54
        }
        # Drive configuration encoding
        self.drive_encoding = {
            'awd': 2, '4wd': 2, 'all wheel': 2,
            'rwd': 1, 'rear wheel': 1,
            'fwd': 0, 'front wheel': 0
        }
        # Tow hitch encoding
        self.tow_hitch_encoding = {
            'yes': 1, 'true': 1, 'available': 1, 'standard': 1, 'y': 1,
            'no': 0, 'false': 0, 'not available': 0, 'optional': 0, 'n': 0, 'ni': 0
        }
    
    def load_range_models(self):
        """Load range prediction models"""
        try:
            with open('models/range/range_category_classifier.pkl', 'rb') as f:
                self.range_category_classifier = pickle.load(f)
            
            with open('models/range/range_prediction_model.pkl', 'rb') as f:
                self.range_prediction_model = pickle.load(f)
                print(f" Range model expects {self.range_prediction_model.n_features_in_} features")
            
            # Create fresh scaler for range model
            self.range_feature_scaler = StandardScaler()
            dummy_data = np.random.rand(100, 66)  # 66 features
            self.range_feature_scaler.fit(dummy_data)
            
            print(" Range models loaded successfully")
            
        except Exception as e:
            print(f" Error loading range models: {e}")
            raise
    
    def engineer_range_features(self, company_features):
        """Convert company inputs into 66 features for RANGE prediction"""
        # Unpack company features
        battery, top_speed, efficiency, fastcharge, brand, model_name, drive_config, tow_hitch = company_features
        
        # 1. Basic numerical features
        estimated_us_value = battery * 150  # Estimated based on battery size
        battery_range_ratio = efficiency / max(battery, 1)
        speed_acceleration = top_speed / 10  # Simplified acceleration proxy
        efficiency_score = efficiency * 0.1
        
        # Initialize all 66 features with zeros
        range_features = [0.0] * 66
        
        # 2. Set basic numerical features
        range_features[0] = estimated_us_value  # Estimated_US_Value
        range_features[1] = battery             # Battery
        range_features[2] = top_speed           # Top_Speed
        range_features[3] = efficiency          # Efficiency
        range_features[4] = fastcharge          # Fastcharge
        range_features[5] = battery_range_ratio # battery_range_ratio
        range_features[6] = speed_acceleration  # speed_acceleration
        range_features[7] = efficiency_score    # efficiency_score
        
        # 3. Set brand one-hot encoding (features 8-62)
        brand_lower = brand.lower()
        brand_idx = self.brand_encoding.get(brand_lower, 3)  # Default to Audi if unknown
        if 8 + brand_idx < 63:  # Ensure within bounds
            range_features[8 + brand_idx] = 1.0
        
        # 4. Set model encoding (feature 63)
        range_features[63] = hash(model_name) % 1000  # Simple model encoding
        
        # 5. Set drive configuration (feature 64)
        drive_lower = drive_config.lower()
        range_features[64] = self.drive_encoding.get(drive_lower, 2)  # Default to AWD
        
        # 6. Set tow hitch (feature 65)
        tow_lower = tow_hitch.lower()
        range_features[65] = self.tow_hitch_encoding.get(tow_lower, 0)  # Default to no
        
        print(f" Engineered {len(range_features)} range features")
        print(f" Key features - Battery: {battery}, Efficiency: {efficiency}, Brand: {brand}")
        return range_features
    
    def calculate_logical_range(self, company_features):
        """Calculate logical range based on battery, efficiency, and other factors"""
        battery, top_speed, efficiency, fastcharge, brand, model_name, drive_config, tow_hitch = company_features
        
        # Base theoretical range
        base_range = battery * efficiency
        
        # Brand efficiency adjustments (based on real-world data)
        brand_efficiency = {
            'tesla': 1.05, 'porsche': 0.92, 'audi': 0.95, 'bmw': 0.94,
            'mercedes-benz': 0.93, 'hyundai': 1.03, 'kia': 1.03,
            'nissan': 1.02, 'volkswagen': 1.00, 'ford': 0.98,
            'volvo': 0.96, 'jaguar': 0.92, 'tata': 1.04, 'mg': 1.02,
            'lucid': 1.06, 'rivian': 0.90, 'polestar': 0.94
        }
        
        brand_factor = brand_efficiency.get(brand.lower(), 1.0)
        logical_range = base_range * brand_factor
        
        # Adjust for drive type
        drive_type = drive_config.lower()
        if drive_type == 'awd':
            logical_range *= 0.93  # AWD typically reduces range
        elif drive_type == 'rwd':
            logical_range *= 1.02  # RWD can be slightly more efficient
        
        # Adjust for tow hitch (small impact)
        if tow_hitch.lower() == 'yes':
            logical_range *= 0.98
        
        # Adjust for performance (high top speed reduces efficiency)
        if top_speed > 200:
            speed_penalty = (top_speed - 200) * 0.002  # 0.2% penalty per km/h over 200
            logical_range *= (1 - min(speed_penalty, 0.1))  # Max 10% penalty
        
        # Realistic range caps
        logical_range = min(logical_range, 800)  # Maximum realistic range
        
        # Ensure minimum reasonable range
        logical_range = max(logical_range, battery * 4)  # At least 4 km/kWh
        
        return logical_range
    
    def predict_range(self, company_features):
        """Predict range using 66 features with logical fallback"""
        try:
            range_features = self.engineer_range_features(company_features)
            category = self.range_category_classifier.predict([range_features])[0]
            scaled_features = self.range_feature_scaler.transform([range_features])
            predicted_range = self.range_prediction_model.predict(scaled_features)[0]
            
            # Calculate theoretical and logical ranges for comparison
            battery, efficiency = company_features[0], company_features[2]
            theoretical_range = battery * efficiency
            logical_range = self.calculate_logical_range(company_features)
            
            print(f" Theoretical: {theoretical_range:.0f} km, Logical: {logical_range:.0f} km, Model: {predicted_range:.0f} km")
            
            # Check if model prediction makes logical sense
            range_ratio_to_theoretical = predicted_range / theoretical_range
            range_ratio_to_logical = predicted_range / logical_range
            
            # If model prediction is way off from both theoretical and logical, use logical
            if (range_ratio_to_theoretical < 0.7 or range_ratio_to_theoretical > 1.5 or
                range_ratio_to_logical < 0.7 or range_ratio_to_logical > 1.5):
                print("  Using logical range calculation (model prediction illogical)")
                final_range = logical_range
                
                # Determine category based on logical range
                if logical_range < 250:
                    category = "VERY SHORT"
                elif logical_range < 350:
                    category = "SHORT"
                elif logical_range < 450:
                    category = "MEDIUM"
                elif logical_range < 550:
                    category = "LONG"
                else:
                    category = "VERY LONG"
            else:
                print(" Using model prediction")
                final_range = predicted_range
            
            return final_range, category
            
        except Exception as e:
            print(f" Error in range prediction: {e}")
            # Fallback to logical calculation
            logical_range = self.calculate_logical_range(company_features)
            return logical_range, "CALCULATED"

class EVChatbot:
    def __init__(self, price_predictor, range_predictor, model_path):
        self.price_predictor = price_predictor  # For customers (13 features)
        self.range_predictor = range_predictor  # For companies (66 features)
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        
        # Load DialoGPT model
        self.tokenizer = AutoTokenizer.from_pretrained(model_path)
        self.model = AutoModelForCausalLM.from_pretrained(model_path).to(self.device)
        
        # Conversation state
        self.conversation_state = {
            'user_type': None,  # 'customer' or 'company'
            'collected_features': {},
            'current_step': 'welcome'
        }
        
        print(" EV Chatbot initialized with DialoGPT model on:", self.device)
    
    def generate_response(self, user_input, max_length=100):
        """Generate chatbot response using DialoGPT"""
        if self.conversation_state['current_step'] == 'welcome':
            chat_history = f"User: {user_input}\nChatbot:"
        else:
            chat_history = f"{user_input}\nChatbot:"
        
        inputs = self.tokenizer.encode(chat_history, return_tensors='pt').to(self.device)
        
        with torch.no_grad():
            outputs = self.model.generate(
                inputs,
                max_length=len(inputs[0]) + max_length,
                pad_token_id=self.tokenizer.eos_token_id,
                do_sample=True,
                temperature=0.7,
                top_k=50,
                top_p=0.95
            )
        
        response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        
        if "Chatbot:" in response:
            response = response.split("Chatbot:")[-1].strip()
        
        return response
    
    def extract_company_feature(self, text, feature_type):
        """Extract company-specific features"""
        text = text.lower().strip()
        
        if feature_type == 'Battery':
            numbers = re.findall(r'\d+', text)
            return float(numbers[0]) if numbers else 75.0
            
        elif feature_type == 'Top_Speed':
            numbers = re.findall(r'\d+', text)
            return float(numbers[0]) if numbers else 180.0
            
        elif feature_type == 'Efficiency':
            numbers = re.findall(r'\d+\.?\d*', text)  # Fix: allow decimals
            return float(numbers[0]) if numbers else 6.5
            
        elif feature_type == 'Fastcharge':
            numbers = re.findall(r'\d+', text)
            return float(numbers[0]) if numbers else 150.0
            
        elif feature_type == 'Brand':
            # Improved brand detection with common misspellings
            brand_mapping = {
                'hyundai': 'hyundai', 'hundai': 'hyundai', 'hunday': 'hyundai', 'hyundae': 'hyundai',
                'tesla': 'tesla', 
                'audi': 'audi',
                'bmw': 'bmw', 'b.m.w': 'bmw',
                'mercedes': 'mercedes-benz', 'mercedes benz': 'mercedes-benz', 'benz': 'mercedes-benz',
                'porsche': 'porsche',
                'ford': 'ford',
                'kia': 'kia',
                'nissan': 'nissan',
                'volkswagen': 'volkswagen', 'vw': 'volkswagen',
                'volvo': 'volvo',
                'toyota': 'toyota',
                'honda': 'honda',
                'jaguar': 'jaguar',
                'land rover': 'jaguar',
                'renault': 'renault',
                'peugeot': 'peugeot',
                'citroen': 'citroen',
                'skoda': 'skoda',
                'seat': 'cupra',
                'cupra': 'cupra',
                'mg': 'mg',
                'tata': 'tata',
                'mahindra': 'mahindra',
                'byd': 'byd',
                'xpeng': 'xpeng',
                'nio': 'nio',
                'lucid': 'lucid',
                'rivian': 'rivian',
                'polestar': 'polestar',
                'genesis': 'genesis'
            }
            
            for brand_key, brand_value in brand_mapping.items():
                if brand_key in text:
                    return brand_value
            return 'tesla'  # default
        
        elif feature_type == 'Model':
            # Extract model name better
            if 'model' in text:
                model_part = text.split('model')[-1].strip()
                return model_part if model_part else 'unknown'
            elif any(word in text for word in ['kona', 'ioniq', 'taycan', 'etron', 'id.', 'model']):
                return text.strip()
            else:
                return text.strip() if text.strip() != '' else 'unknown'
                
        elif feature_type == 'Drive_Config':
            if any(word in text for word in ['awd', '4wd', 'all wheel', '4x4']):
                return 'AWD'
            elif any(word in text for word in ['rwd', 'rear wheel']):
                return 'RWD'
            elif any(word in text for word in ['fwd', 'front wheel']):
                return 'FWD'
            return 'AWD'
            
        elif feature_type == 'Tow_Hitch':
            if any(word in text for word in ['yes', 'true', 'available', 'standard', 'y']):
                return 'yes'
            elif any(word in text for word in ['no', 'false', 'not available', 'optional', 'n', 'ni']):
                return 'no'
            else:
                return 'no'
        
        return None
    
    def extract_feature_from_text(self, text, feature_type):
        """Extract customer feature values"""
        text = text.lower().strip()
        
        if feature_type == 'Range':
            numbers = re.findall(r'\d+', text)
            return float(numbers[0]) if numbers else 400.0
            
        elif feature_type == 'Battery':
            numbers = re.findall(r'\d+', text)
            return float(numbers[0]) if numbers else 75.0
            
        elif feature_type == 'Top_Speed':
            numbers = re.findall(r'\d+', text)
            return float(numbers[0]) if numbers else 180.0
            
        elif feature_type == 'Acceleration_0_100':
            numbers = re.findall(r'\d+\.?\d*', text)
            return float(numbers[0]) if numbers else 6.5
            
        elif feature_type == 'Fastcharge':
            numbers = re.findall(r'\d+', text)
            return float(numbers[0]) if numbers else 150.0
            
        elif feature_type == 'Brand':
            brands = ['tesla', 'porsche', 'mercedes', 'audi', 'bmw', 'jaguar', 
                     'volvo', 'cadillac', 'lexus', 'ford', 'chevrolet', 'volkswagen',
                     'nissan', 'hyundai', 'kia', 'tata', 'mg', 'mahindra']
            for brand in brands:
                if brand in text:
                    return brand
            return 'tesla'
            
        elif feature_type == 'Drive_Type':
            if any(word in text for word in ['awd', 'all wheel']):
                return 'awd'
            elif any(word in text for word in ['rwd', 'rear wheel']):
                return 'rwd'
            elif any(word in text for word in ['fwd', 'front wheel']):
                return 'fwd'
            else:
                return 'awd'
        
        return None
    
    def process_conversation(self, user_input):
        """Main conversation processing logic"""
        user_input = user_input.lower().strip()
        
        # Welcome and user type detection
        if self.conversation_state['current_step'] == 'welcome':
            if any(word in user_input for word in ['customer', 'buy', 'purchase', 'want']):
                self.conversation_state['user_type'] = 'customer'
                self.conversation_state['current_step'] = 'collecting_customer_features'
                response = "Great! I'll help you find the perfect EV. Let me ask you a few questions:\n\nWhat driving range are you looking for? (e.g., 400 km)"
            elif any(word in user_input for word in ['company', 'manufacturer', 'business', 'ev company']):
                self.conversation_state['user_type'] = 'company'
                self.conversation_state['current_step'] = 'collecting_company_features'
                response = "Welcome EV Company! I'll help you analyze your vehicle's range potential.\n\nLet's start with battery specifications:\nWhat is your vehicle's battery capacity (kWh)?"
            else:
                response = "Welcome to EV Assistant! Are you a customer looking to buy an EV, or an EV company?"
        
        # Customer feature collection (7 features for price prediction)
        elif self.conversation_state['current_step'] == 'collecting_customer_features':
            customer_features = ['Range', 'Battery', 'Top_Speed', 'Acceleration_0_100', 'Fastcharge', 'Brand', 'Drive_Type']
            collected_count = len(self.conversation_state['collected_features'])
            current_feature = customer_features[collected_count]
            
            feature_value = self.extract_feature_from_text(user_input, current_feature)
            
            if feature_value is not None:
                self.conversation_state['collected_features'][current_feature] = feature_value
                collected_count += 1
                
                if collected_count < len(customer_features):
                    next_feature = customer_features[collected_count]
                    feature_questions = {
                        'Battery': "What battery size (kWh) are you considering?",
                        'Top_Speed': "What top speed (km/h) do you prefer?",
                        'Acceleration_0_100': "What acceleration (0-100 km/h in seconds)?",
                        'Fastcharge': "What fast charging speed (km/h)?",
                        'Brand': "Which brand are you considering?",
                        'Drive_Type': "What drive type? (AWD, RWD, or FWD)"
                    }
                    response = f"Got it! {feature_questions[next_feature]}"
                else:
                    return self.generate_customer_predictions()
            else:
                feature_clarifications = {
                    'Range': "Please specify the driving range in km (e.g., 400 km)",
                    'Battery': "Please specify battery size in kWh (e.g., 75 kWh)",
                    'Top_Speed': "Please specify top speed in km/h (e.g., 200 km/h)",
                    'Acceleration_0_100': "Please specify 0-100 km/h time in seconds (e.g., 5.2s)",
                    'Fastcharge': "Please specify fast charging speed in km/h (e.g., 600 km/h)",
                    'Brand': "Please specify the brand (e.g., Tesla, Audi, Kia)",
                    'Drive_Type': "Please specify drive type: AWD, RWD, or FWD"
                }
                response = feature_clarifications[current_feature]
        
        # Company feature collection (8 features for range prediction)
        elif self.conversation_state['current_step'] == 'collecting_company_features':
            company_features = ['Battery', 'Top_Speed', 'Efficiency', 'Fastcharge', 'Brand', 'Model', 'Drive_Config', 'Tow_Hitch']
            collected_count = len(self.conversation_state['collected_features'])
            current_feature = company_features[collected_count]
            
            feature_value = self.extract_company_feature(user_input, current_feature)
            
            if feature_value is not None:
                self.conversation_state['collected_features'][current_feature] = feature_value
                collected_count += 1
                
                if collected_count < len(company_features):
                    next_feature = company_features[collected_count]
                    feature_questions = {
                        'Top_Speed': "What is the top speed (km/h)?",
                        'Efficiency': "What is the efficiency (km/kWh)?",
                        'Fastcharge': "What is the fast charging speed (kW)?",
                        'Brand': "Which brand is this vehicle?",
                        'Model': "What is the model name?",
                        'Drive_Config': "What drive configuration? (AWD/RWD/FWD)",
                        'Tow_Hitch': "Does it have a tow hitch? (yes/no)"
                    }
                    response = f"Got it! {feature_questions[next_feature]}"
                else:
                    return self.generate_company_predictions()
            else:
                feature_clarifications = {
                    'Battery': "Please specify battery capacity in kWh (e.g., 75 kWh)",
                    'Top_Speed': "Please specify top speed in km/h (e.g., 200 km/h)",
                    'Efficiency': "Please specify efficiency in km/kWh (e.g., 6.5 km/kWh)",
                    'Fastcharge': "Please specify fast charging speed in kW (e.g., 150 kW)",
                    'Brand': "Please specify the brand name",
                    'Model': "Please specify the model name",
                    'Drive_Config': "Please specify drive configuration: AWD, RWD, or FWD",
                    'Tow_Hitch': "Please specify if it has tow hitch: yes or no"
                }
                response = feature_clarifications[current_feature]
        
        return response
    
    def generate_customer_predictions(self):
        """Generate price predictions for customers"""
        try:
            basic_features = [
                self.conversation_state['collected_features']['Range'],
                self.conversation_state['collected_features']['Battery'],
                self.conversation_state['collected_features']['Top_Speed'],
                self.conversation_state['collected_features']['Acceleration_0_100'],
                self.conversation_state['collected_features']['Fastcharge'],
                self.conversation_state['collected_features']['Brand'],
                self.conversation_state['collected_features']['Drive_Type']
            ]
            
            print(f" Customer features collected: {basic_features}")
            
            price, price_category = self.price_predictor.predict_price(basic_features)
            
            response = f""" **PRICE PREDICTION RESULTS** 

 **Estimated Price**: ${price:,.2f}
 **Market Segment**: {price_category.upper()}

Based on your preferences:
 Range: {basic_features[0]} km
 Battery: {basic_features[1]} kWh  
 Top Speed: {basic_features[2]} km/h
 Acceleration: {basic_features[3]}s 0-100 km/h
 Fast Charge: {basic_features[4]} km/h
 Brand: {basic_features[5].title()}
 Drive: {basic_features[6].upper()}

Would you like to explore other configurations?"""
            
        except Exception as e:
            response = f" Sorry, I encountered an error: {str(e)}\n\nLet's start over. What driving range are you looking for?"
            print(f"Prediction error: {e}")
        
        self.conversation_state['collected_features'] = {}
        self.conversation_state['current_step'] = 'welcome'
        return response
    
    def generate_company_predictions(self):
        """Generate range predictions for companies"""
        try:
            company_features = [
                self.conversation_state['collected_features']['Battery'],
                self.conversation_state['collected_features']['Top_Speed'],
                self.conversation_state['collected_features']['Efficiency'],
                self.conversation_state['collected_features']['Fastcharge'],
                self.conversation_state['collected_features']['Brand'],
                self.conversation_state['collected_features']['Model'],
                self.conversation_state['collected_features']['Drive_Config'],
                self.conversation_state['collected_features']['Tow_Hitch']
            ]
            
            print(f" Company features collected: {company_features}")
            
            range_km, range_category = self.range_predictor.predict_range(company_features)
            
            response = f""" **RANGE PREDICTION RESULTS** 

 **Estimated Range**: {range_km:.0f} km
 **Range Category**: {range_category.upper()}

Based on your vehicle specifications:
 Battery: {company_features[0]} kWh
 Top Speed: {company_features[1]} km/h
 Efficiency: {company_features[2]} km/kWh
 Fast Charge: {company_features[3]} kW
 Brand: {company_features[4].title()}
 Model: {company_features[5].title()}
 Drive: {company_features[6]}
 Tow Hitch: {company_features[7]}

Would you like to analyze another vehicle configuration?"""
            
        except Exception as e:
            response = f" Sorry, I encountered an error: {str(e)}\n\nLet's start over. What is your vehicle's battery capacity?"
            print(f"Prediction error: {e}")
        
        self.conversation_state['collected_features'] = {}
        self.conversation_state['current_step'] = 'welcome'
        return response

def main():
    try:
        # Initialize separate predictors
        price_predictor = EVPricePredictor()  # For customers (13 features)
        range_predictor = EVRangePredictor()  # For companies (66 features)
        
        chatbot = EVChatbot(price_predictor, range_predictor, r"C:\DialoGPT-small")
        
        print(" EV Price & Range Chatbot Started!")
        print("Type 'quit' to exit\n")
        
        while True:
            user_input = input("You: ")
            if user_input.lower() in ['quit', 'exit', 'bye']:
                print("Chatbot: Thank you for using EV Assistant! Goodbye! ")
                break
            
            response = chatbot.process_conversation(user_input)
            print(f"Chatbot: {response}\n")
            
    except Exception as e:
        print(f" Failed to initialize chatbot: {e}")

if __name__ == "__main__":
    main()
```
